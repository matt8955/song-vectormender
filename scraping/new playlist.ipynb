```python
import pandas as pd
import numpy as np
from tqdm import tqdm
tqdm.pandas()
import spotipy
```

```python
import os

import re


def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = os.listdir(path_to_dir)
    return [ filename for filename in filenames if 
            (filename.endswith(suffix) and filename.startswith('spotify_playlist'))]

def get_playlist_id(row):
    regex = re.findall(r'\bplaylist\b\/\w*', row)
    if regex:
        return regex[0][9:]
    else:
        return np.nan

csvs = find_csv_filenames(os.getcwd())
csvs.sort(reverse=False)

df_list = [pd.read_csv(x, usecols=['tweet','urls']) for x in csvs]

df = pd.concat(df_list)

df['playlist_id'] = df.tweet.apply(lambda x: get_playlist_id(x))

df.dropna(subset=["playlist_id"],inplace=True)

df.drop_duplicates(subset=["playlist_id"],inplace=True)
unique_playlists = list(df.playlist_id)

from spotipy.oauth2 import SpotifyClientCredentials

client_id='2264c5e5d4c1469d812c8097babe7e42'
client_secret='2050b324730b46aea17caa8e51175da9'

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
spot = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

def get_playlist_df(playlist_id):
    try:
        result = spot.user_playlist(user = None, playlist_id= playlist_id)
        playlist_name = result['name']
        tracks = result['tracks']['items']
        tracklist = []
        for i, track in enumerate(tracks):
            try:
                tid = track['track']['id']
                pop = track['track']['popularity']
                tracklist.append((i+1,tid, pop))
            except:
                tracklist.append((np.nan,np.nan,np.nan))
        df = pd.DataFrame(tracklist)
        df['playlist_id'] = playlist_id
        df['playlist_name'] = result['name']
        df['user_id'] = result['owner']['id']
        try:
            df['followers'] = result['followers']['total']
        except:
            np.nan
#         x = ['track_id','song_id','popularity', 'playlist_id','playlist_name', 'user_id', 'followers']    
#         df.columns = x
#         df.reset_index()
        return (playlist_id,df)
    
    except spotipy.SpotifyException:
        return (0, playlist_id)
```
