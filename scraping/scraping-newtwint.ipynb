```python
date = '2019-11-18'
while date != '2019-06-10':
    c = twint.Config()
    c.Search = 'open.spotify.com/playlist/'
    c.Output = f"spotify_playlist_tweets3.csv"
    c.Since = '2019-06-10'
    c.Until = '2019-11-22'
    c.Lang = 'en'
    c.Hide_output = True
    c.Store_csv = True
    c.Debug = True
    c.Resume = "twint-last-request.log"
    df = pd.read_csv(f"spotify_playlist_tweets3.csv")
    date = df.iloc[-1].date
    twint.run.Search(c)
```

```python
since = '2018-06-10'
until = '2019-06-10'
date = '0'
i=0
while date != since:
    c = twint.Config()
    c.Search = 'open.spotify.com/playlist/'
    c.Output = f"spotify_playlist_tweets4.csv"
    c.Since = since
    c.Until = until
    c.Lang = 'en'
    c.Hide_output = True
    c.Store_csv = True
    c.Debug = True
    if i>0:
        c.Resume = "twint-last-request.log"
    twint.run.Search(c)
    df = pd.read_csv(f"spotify_playlist_tweets4.csv")
    date = df.iloc[-1].date
    twint.run.Search(c)
    i = i+1
    print(f'completed run {i}')
```
